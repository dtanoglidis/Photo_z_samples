### Horizontal_FoM_bars

```python
# Import stuff
import numpy as np
import scipy
import matplotlib
import matplotlib.pyplot as plt
from scipy import interpolate 
from code.plot_utils import plot_pretty
from matplotlib import rcParams
rcParams['font.family'] = 'serif' 
%matplotlib inline
%pylab inline 
```

```python
FoM_auto_free = 4126.144975978385
FoM_auto_cons = 13707.131778968247
FoM_auto_optim = 49905.38039846701
FoM_auto_fixed = 86304.63038448585

FoM_cross_free = 4301.6521634393075
FoM_cross_cons = 14062.4747311969
FoM_cross_optim = 50965.38493857575
FoM_cross_fixed = 88585.73651285711
```

```python
data_auto = [1.0, FoM_auto_cons/FoM_auto_free,  FoM_auto_optim/FoM_auto_free,  FoM_auto_fixed/FoM_auto_free]
data_cross = [FoM_cross_free/FoM_auto_free, FoM_cross_cons/FoM_auto_free, FoM_cross_optim/FoM_auto_free, FoM_cross_fixed/FoM_auto_free]

x = [0,2,4,6]
x_2 = [-0.5, 1.5, 3.5, 5.5]

plot_pretty()
fig = plt.figure(figsize = (4,3.5))

plt.barh(x,data_auto, height =0.5, color='navy', label='Auto only')
plt.barh(x_2,data_cross, height =0.5, color='firebrick', label='Auto + Cross')
plt.grid(ls='--', axis='x', alpha=0.7)

plt.xticks([0,5,10,15,20], fontsize=12)
plt.yticks([-0.2,1.8,3.8,5.8], ('Free', 'Conservative\nPrior', 'Optimistic\nPrior', 'Fixed'), fontsize=12)

plt.xlabel('FoM (normalized)', fontsize=13)

plt.legend(frameon=True, loc='lower right', fontsize=13)
tight_layout()

plt.savefig('Relative_FoMs.pdf')
plt.show()
```
